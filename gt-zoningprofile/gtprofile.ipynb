## Garnet Zoning Profile with Pandas

First load the modules and load the datafile in a pandas dataframe. The csv-File is not comma-separated but uses tabulator instead. Show the dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data.txt', delimiter='\t')
df
```

```python
plt.style.use('fivethirtyeight')

fig, ax1 = plt.subplots()
ax1.set_xlabel('x/mm')
ax1.set_ylabel('Xalm, Xgrs, Xprp, Xsps')
ax1.plot(df['Xmm'], df['Xal'], marker='D', markersize=3, markeredgecolor='black', linewidth=1, color='green', label='Xalm')
ax1.plot(df['Xmm'], df['Xsp'], color='orange', label='Xsps')
ax1.plot(df['Xmm'], df['Xgr'], color='y', label='Xgrs')
ax1.plot(df['Xmm'], df['Xpr'], color='blue', label='Xprp')
ax1.set_xlim([0, 2.4])

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'red'
ax2.set_ylabel('XFe')
y2start=0.7
y2end=0.9
ax2.set_ylim([y2start, y2end])
ax2.yaxis.set_ticks(np.arange(y2start, y2end, .05))

ax2.minorticks_on()
ax2.plot(df['Xmm'], df['XFe'], color=color, label='XFe')

# ask matplotlib for the plotted objects and their labels
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
#ax1.legend(lines + lines2, labels + labels2, loc=0)
# Place legend outside of diagram
fig.legend(lines + lines2, labels + labels2, loc='upper right')

plt.title('Garnet Profile')
plt.savefig('profile.pdf')
plt.show()
```

```python
import seaborn as sns
fig, ax1 = plt.subplots() # initializes figure and plots
ax1.set_xlim([0, 2.4])
sns.set_style('darkgrid')
ax2 = ax1.twinx()
ax2.grid(visible=False)
sns.lineplot( x='Xmm', y='Xsp', data= df, ax=ax1)
sns.lineplot( x='Xmm', y='Xgr', data= df, ax=ax1)
sns.lineplot( x='Xmm', y='Xpr', data= df, ax=ax1)
sns.lineplot( x='Xmm', y='XFe', data= df, ax=ax2)


plt.show()
```
